```python
import pandas as pd

file_path = 'TikTok_songs_2022.csv'
df = pd.read_csv(file_path)

print("First 5 rows of the dataset:")
df.head()
```

```python
print("\nDataset Information:")
df.info()
```

```python
# Sort songs by popularity(track_pop)
top_songs_df = df.sort_values(by='track_pop', ascending=False)

# Top 30 songs frame
top_30_songs = top_songs_df.head(30)

# Print top 30 songs
print("Top 30 Most Popular TikTok Songs of 2022:")
top_30_songs[['track_name', 'artist_name', 'album', 'track_pop']]
```
